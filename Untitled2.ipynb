## **Tugas Praktikum**

```python
import pandas as pd

df5 = pd.read_csv('/content/jobsheet2-tugas/wbc.csv')
df5.head()
```

## 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df5.isnull().sum()
df5 = df5[[
  # 'id',
  'diagnosis',
  'radius_mean',
  'texture_mean',
  'perimeter_mean',
  'area_mean',
  'smoothness_mean',
  'compactness_mean',
  'concavity_mean',
  'concave points_mean',
  'symmetry_mean',
  'fractal_dimension_mean',
  'radius_se',
  'texture_se',
  'perimeter_se',
  'area_se',
  'smoothness_se',
  'compactness_se',
  'concavity_se',
  'concave points_se',
  'symmetry_se',
  'fractal_dimension_se',
  'radius_worst',
  'texture_worst',
  'perimeter_worst',
  'area_worst',
  'smoothness_worst',
  'compactness_worst',
  'concavity_worst',
  'concave points_worst',
  'symmetry_worst',
  'fractal_dimension_worst',
]]
df5.info()
df5.head()
```

## 2. Lakukan proses encoding pada kolom "diagnosis".

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()
df5['diagnosis'] = le.fit_transform(df5['diagnosis'])
df5.head()
```

## 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
df5['radius_mean'] = std.fit_transform(df5[['radius_mean']])
df5['texture_mean'] = std.fit_transform(df5[['texture_mean']])
df5['perimeter_mean'] = std.fit_transform(df5[['perimeter_mean']])
df5['area_mean'] = std.fit_transform(df5[['area_mean']])
df5['radius_se'] = std.fit_transform(df5[['radius_se']])
df5['texture_se'] = std.fit_transform(df5[['texture_se']])
df5['perimeter_se'] = std.fit_transform(df5[['perimeter_se']])
df5['area_se'] = std.fit_transform(df5[['area_se']])
df5['radius_worst'] = std.fit_transform(df5[['radius_worst']])
df5['texture_worst'] = std.fit_transform(df5[['texture_worst']])
df5['perimeter_worst'] = std.fit_transform(df5[['perimeter_worst']])
df5['area_worst'] = std.fit_transform(df5[['area_worst']])
df5.head(20)
```

## 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
from sklearn.model_selection import train_test_split
# Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20

df5_train, df5_unsee = train_test_split(df5, test_size=0.2, random_state=0, stratify=df5['diagnosis'])
df5_val, df5_test = train_test_split(df5_unsee, test_size=0.5, random_state=0, stratify=df5_unsee['diagnosis'])
print(f'Jumlah data asli: {df5.shape[0]}')
print(f'Jumlah data train: {df5_train.shape[0]}')
print(f'Jumlah data unsee: {df5_unsee.shape[0]}')
print(f'Jumlah data val: {df5_val.shape[0]}')
print(f'Jumlah data test: {df5_test.shape[0]}')
# Cek rasio tiap label
print('=========')
print(f'Jumlah label data asli:\n{df5["diagnosis"].value_counts()}')
print(f'Jumlah label data train:\n{df5_train["diagnosis"].value_counts()}')
print(f'Jumlah label data val:\n{df5_val["diagnosis"].value_counts()}')
print(f'Jumlah label data test:\n{df5_test["diagnosis"].value_counts()}')
```
